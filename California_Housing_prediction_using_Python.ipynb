# **Califorina Housing Prediction**

### **Introduction**

The purpose of the project is to predict median house values in Californian districts, given many features from these districts.

#### Background of Problem Statement :

The US Census Bureau has published California Census Data which has 10 types of metrics such as the population, median income, median housing price, and so on for each block group in California. The dataset also serves as an input for project scoping and tries to specify the functional and nonfunctional requirements for it.

1. longitude: A measure of how far west a house is; a higher value is farther west

2. latitude: A measure of how far north a house is; a higher value is farther north

3. housingMedianAge: Median age of a house within a block; a lower number is a newer building

4. totalRooms: Total number of rooms within a block

5. totalBedrooms: Total number of bedrooms within a block

6. population: Total number of people residing within a block

7. households: Total number of households, a group of people residing within a home unit, for a block

8. medianIncome: Median income for households within a block of houses (measured in tens of thousands of US Dollars)

9. medianHouseValue: Median house value for households within a block (measured in US Dollars)

10. oceanProximity: Location of the house w.r.t ocean/sea

## **Classification of Data**


1.   longitude - It is Continuous data
2.   latitude  - It is Continuous Data
3.   housing_median_age - - It is Continuous data
4.   total rooms - - It is discrete data
5.   total bedrooms - It is discrete data
6.   Population - It is Continuous data
7.   households - It is Continuous data
8.   median income - It is Continuous data
9.   median house value - It is discrete data
10.   ocean proximity - It is nominal data

# **Question 10**

Please create a new column named total_bedroom _size .
if the size is 10 or less, it should be quoted as small. If the size is 11 or more but less than 1000, it should be median, otherwise, it should be considered large.

```python
# this is my orginal dataset

df.head(5)
```

```python
# to add new column firstly we should know the min and max total_bedrooms

df["total_bedrooms"].max() 
```

```python
df["total_bedrooms"].min() 
```

```python
def totalBedroomSize(column):
    if column <= 10:
        return "small"
    elif column >= 11 and column <1000:
        return "mediun"
    else:
        return "large"
```

```python
# Adding a new column names as total_bedroom_size. to tell whether the room size is small, mediam or large

df["total_bedroom_size"] = df["total_bedrooms"].apply(totalBedroomSize)
df
```
